# Phase 4: ESPN Points Mapper
Calculate ESPN H2H fantasy points — the ML target variable.

```python
import sys
sys.path.insert(0, '..')

from src.espn_points_mapper import build_espn_points
from src.data_builder import load_config

cfg = load_config()
batters, pitchers = build_espn_points(cfg)

qual_b = batters[~batters['LOW_SAMPLE']]
qual_p = pitchers[~pitchers['LOW_SAMPLE']]
print(f'Batters: {len(qual_b)} qualified rows')
print(f'Pitchers: {len(qual_p)} qualified rows')
```

```python
# Top producers by year
for y in [2021, 2022, 2023, 2024]:
    top = qual_b[qual_b['year']==y].nlargest(3, 'ESPN_Pts')
    print(f'{y}: {list(zip(top["last"], top["ESPN_Pts"].astype(int)))}')
```

```python
# Pitcher top producers
for y in [2021, 2022, 2023, 2024]:
    top = qual_p[qual_p['year']==y].nlargest(3, 'ESPN_Pts')
    print(f'{y}: {list(zip(top["last"], top["ESPN_Pts"].astype(int)))}')
```
