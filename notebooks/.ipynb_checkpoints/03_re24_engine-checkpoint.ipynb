```python
import sys
sys.path.insert(0, '..')

from src.re24_engine import build_re24_outputs
from src.data_builder import load_config

cfg = load_config()
results = build_re24_outputs(cfg)
```

```python
# Verify RE matrix
re = results['re_matrix']
print('RE Matrix (2023):')
pivot = re[re['year']==2023].pivot(index='base_state', columns='outs', values='avg_re')
print(pivot.round(3))
print(f'\nBases empty, 0 outs: {pivot.loc[0, 0]:.3f} (expected ~0.50)')
```

```python
# wOBA weights
sc = results['season_constants']
print(sc[['year','wBB','w1B','w2B','w3B','wHR']].to_string(index=False))
```

```python
# Top RE24 batters 2023
bre = results['batter_re24']
print(bre[bre['year']==2023].nlargest(10, 'RE24')[['id','RE24','PA']].to_string(index=False))
```
