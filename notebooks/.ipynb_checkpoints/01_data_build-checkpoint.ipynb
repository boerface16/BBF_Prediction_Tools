# Phase 1: Data Build
Aggregate game-level Retrosheet data to season-level batting and pitching stats.

```python
import sys
sys.path.insert(0, '..')

from src.data_builder import load_config, build_batters, build_pitchers

cfg = load_config()
print('Config loaded.')
```

```python
# Build season-level batting
batters = build_batters(cfg)
print(f'Batters shape: {batters.shape}')
print(f'Years: {sorted(batters["year"].unique())}')
print(f'\nBatters per year (all):')
print(batters.groupby('year').size())
print(f'\nBatters per year (qualified):')
print(batters[~batters['LOW_SAMPLE']].groupby('year').size())
print(f'\nSample row:')
batters.head(3)
```

```python
# Build season-level pitching
pitchers = build_pitchers(cfg)
print(f'Pitchers shape: {pitchers.shape}')
print(f'Years: {sorted(pitchers["year"].unique())}')
print(f'\nPitchers per year (all):')
print(pitchers.groupby('year').size())
print(f'\nPitchers per year (qualified):')
print(pitchers[~pitchers['LOW_SAMPLE']].groupby('year').size())
print(f'\nSample row:')
pitchers.head(3)
```

```python
# Verification: W+L per year should approximate total games * 2 (one W and one L per game)
# Total regular season games ~ 2430, so W+L ~ 4860
wl_check = pitchers.groupby('year')[['W', 'L']].sum()
wl_check['W+L'] = wl_check['W'] + wl_check['L']
print('W/L/SV verification:')
print(wl_check)
print(f'\nTotal SV per year:')
print(pitchers.groupby('year')['SV'].sum())
```

```python
# Spot check: verify a known player
# Look for a high-profile batter
star = batters[batters['last'].str.contains('Ohtani', na=False)]
if len(star) > 0:
    print('Ohtani batting seasons:')
    print(star[['year', 'b_pa', 'b_hr', 'b_r', 'b_rbi', 'b_sb', 'G', 'primary_pos']].to_string(index=False))
else:
    star = batters[batters['last'].str.contains('Judge', na=False)]
    print('Judge batting seasons:')
    print(star[['year', 'b_pa', 'b_hr', 'b_r', 'b_rbi', 'b_sb', 'G', 'primary_pos']].to_string(index=False))
```

```python
# Check dtypes and nulls
print('Batter nulls:')
print(batters.isnull().sum()[batters.isnull().sum() > 0])
print(f'\nPitcher nulls:')
print(pitchers.isnull().sum()[pitchers.isnull().sum() > 0])
print(f'\nBatter dtypes:')
print(batters.dtypes)
```
