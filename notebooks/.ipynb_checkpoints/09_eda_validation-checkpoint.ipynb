# EDA Validation: Compare Our Stats vs FanGraphs/Statcast
Join our derived stats against external "truth" data to catch calculation errors.

```python
import pandas as pd
import numpy as np
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt
import seaborn as sns
import yaml
import os
from pathlib import Path

# Navigate to project root regardless of where kernel starts
if not Path('config.yaml').exists() and Path('../config.yaml').exists():
    os.chdir('..')

with open('config.yaml') as f:
    cfg = yaml.safe_load(f)

# --- Load all CSVs ---
batters_season = pd.read_csv(cfg['paths']['processed']['batters_season'])
pitchers_season = pd.read_csv(cfg['paths']['processed']['pitchers_season'])
batter_features = pd.read_csv(cfg['paths']['processed']['batter_features'])
pitcher_features = pd.read_csv(cfg['paths']['processed']['pitcher_features'])

fg_bat = pd.read_csv('data/raw/features/fangraphs_batters_leaderboard_2015_2025.csv')
fg_pit = pd.read_csv('data/raw/features/fangraphs_pitchers_leaderboard_2015_2025.csv')
sc_pit = pd.read_csv('data/raw/features/statcast_pitchers_leaderboard_2015_2025.csv')
id_map = pd.read_csv('data/raw/features/id_map_2015_2025.csv')

# Filter out bad id_map rows
id_map = id_map[id_map['key_fangraphs'] > 0].copy()

print(f'batters_season: {batters_season.shape}')
print(f'pitchers_season: {pitchers_season.shape}')
print(f'batter_features: {batter_features.shape}')
print(f'pitcher_features: {pitcher_features.shape}')
print(f'fg_bat: {fg_bat.shape}')
print(f'fg_pit: {fg_pit.shape}')
print(f'sc_pit: {sc_pit.shape}')
print(f'id_map: {id_map.shape}')

# --- Helpers ---
def parse_fg_pct(s):
    """Parse FG pct column: '25.3 %' -> 0.253, or pass through if already float."""
    if isinstance(s, str):
        return float(s.replace('%', '').strip()) / 100
    return s / 100 if s > 1 else s


ALL_SUMMARIES = []  # collect summary rows across cells


def compare_stats(merged_df, pairs, stat_type='rate', source_label=''):
    """
    pairs: list of (ours_col, truth_col, label)
    stat_type: 'rate' (threshold 0.01) or 'index' (threshold 1.0)
    Returns summary DataFrame.
    """
    threshold = 0.01 if stat_type == 'rate' else 1.0
    rows = []
    for ours_col, truth_col, label in pairs:
        sub = merged_df[[ours_col, truth_col]].dropna()
        if len(sub) == 0:
            print(f'  {label}: NO MATCHED ROWS')
            rows.append({'stat': label, 'source': source_label, 'mean_diff': np.nan,
                         'mean_abs_diff': np.nan, 'max_abs_diff': np.nan,
                         'correlation': np.nan, 'n_matched': 0, 'n_flagged': 0,
                         'pass_fail': 'NO DATA'})
            continue
        diff = sub[ours_col] - sub[truth_col]
        abs_diff = diff.abs()
        mean_diff = diff.mean()
        mean_abs = abs_diff.mean()
        max_abs = abs_diff.max()
        corr = sub[ours_col].corr(sub[truth_col])
        n_flagged = int((abs_diff > threshold).sum())
        passed = mean_abs <= threshold and corr >= 0.95
        status = 'PASS' if passed else 'FAIL'
        print(f'  {label:25s} mean_abs={mean_abs:.5f}  max_abs={max_abs:.5f}  corr={corr:.4f}  n={len(sub):5d}  flagged={n_flagged:4d}  {status}')
        rows.append({'stat': label, 'source': source_label, 'mean_diff': round(mean_diff, 6),
                     'mean_abs_diff': round(mean_abs, 6), 'max_abs_diff': round(max_abs, 6),
                     'correlation': round(corr, 4), 'n_matched': len(sub),
                     'n_flagged': n_flagged, 'pass_fail': status})
    summary = pd.DataFrame(rows)
    ALL_SUMMARIES.append(summary)
    return summary


# --- Qualification filter helper ---
def apply_qual(df, player_type='batter'):
    """Apply PA/IP qualification filters, relaxed for 2020."""
    if player_type == 'batter':
        pa_col = 'b_pa'
        normal_thresh = cfg['thresholds']['batter_pa']
        covid_thresh = cfg['thresholds']['batter_pa_2020']
    else:
        pa_col = 'p_ipouts'
        normal_thresh = cfg['thresholds']['pitcher_ip_outs']
        covid_thresh = cfg['thresholds']['pitcher_ip_outs_2020']
    mask = np.where(df['year'] == 2020, df[pa_col] >= covid_thresh, df[pa_col] >= normal_thresh)
    return df[mask].copy()


# --- Build merge frames ---
# FG joins: our id -> id_map key_retro -> key_fangraphs -> FG IDfg + year=Season
idm = id_map[['key_retro', 'key_fangraphs', 'key_mlbam']].drop_duplicates()

# Batter season + FG
batter_fg_merge = (batters_season
    .merge(idm, left_on='id', right_on='key_retro', how='inner')
    .merge(fg_bat, left_on=['key_fangraphs', 'year'], right_on=['IDfg', 'Season'], how='inner',
           suffixes=('', '_fg')))
batter_fg_merge = apply_qual(batter_fg_merge, 'batter')

# Pitcher season + FG
pitcher_fg_merge = (pitchers_season
    .merge(idm, left_on='id', right_on='key_retro', how='inner')
    .merge(fg_pit, left_on=['key_fangraphs', 'year'], right_on=['IDfg', 'Season'], how='inner',
           suffixes=('', '_fg')))
pitcher_fg_merge = apply_qual(pitcher_fg_merge, 'pitcher')

# Batter features + FG
batter_feat_fg_merge = (batter_features
    .merge(idm, left_on='id', right_on='key_retro', how='inner')
    .merge(fg_bat, left_on=['key_fangraphs', 'year'], right_on=['IDfg', 'Season'], how='inner',
           suffixes=('', '_fg')))
batter_feat_fg_merge = apply_qual(batter_feat_fg_merge, 'batter')

# Pitcher features + FG
pitcher_feat_fg_merge = (pitcher_features
    .merge(idm, left_on='id', right_on='key_retro', how='inner')
    .merge(fg_pit, left_on=['key_fangraphs', 'year'], right_on=['IDfg', 'Season'], how='inner',
           suffixes=('', '_fg')))
pitcher_feat_fg_merge = apply_qual(pitcher_feat_fg_merge, 'pitcher')

# Pitcher features + Statcast LB
pitcher_feat_sc_merge = (pitcher_features
    .merge(idm, left_on='id', right_on='key_retro', how='inner')
    .merge(sc_pit, left_on=['key_mlbam', 'year'], right_on=['player_id', 'year'], how='inner',
           suffixes=('', '_sc')))
pitcher_feat_sc_merge = apply_qual(pitcher_feat_sc_merge, 'pitcher')

# Match rates
print(f'\n--- Match rates ---')
for name, merged, total in [
    ('batter_fg_merge', batter_fg_merge, len(apply_qual(batters_season, 'batter'))),
    ('pitcher_fg_merge', pitcher_fg_merge, len(apply_qual(pitchers_season, 'pitcher'))),
    ('batter_feat_fg_merge', batter_feat_fg_merge, len(apply_qual(batter_features, 'batter'))),
    ('pitcher_feat_fg_merge', pitcher_feat_fg_merge, len(apply_qual(pitcher_features, 'pitcher'))),
    ('pitcher_feat_sc_merge', pitcher_feat_sc_merge, len(apply_qual(pitcher_features, 'pitcher'))),
]:
    rate = len(merged) / total * 100 if total > 0 else 0
    print(f'  {name:30s} {len(merged):5d} / {total:5d} = {rate:.1f}%')
```

```python
# Cell 2: Batter Season Stats vs FG
print('=== Batter Season Stats vs FanGraphs ===')

# Parse FG pct columns
batter_fg_merge['fg_K_pct'] = batter_fg_merge['K%'].apply(parse_fg_pct)
batter_fg_merge['fg_BB_pct'] = batter_fg_merge['BB%'].apply(parse_fg_pct)

bat_season_pairs = [
    ('AVG', 'AVG_fg', 'AVG'),
    ('OBP', 'OBP_fg', 'OBP'),
    ('SLG', 'SLG_fg', 'SLG'),
    ('OPS', 'OPS_fg', 'OPS'),
    ('ISO', 'ISO_fg', 'ISO'),
    ('BABIP', 'BABIP_fg', 'BABIP'),
    ('K_pct', 'fg_K_pct', 'K_pct'),
    ('BB_pct', 'fg_BB_pct', 'BB_pct'),
]

bat_season_summary = compare_stats(batter_fg_merge, bat_season_pairs, 'rate', 'bat_season_vs_FG')
```

```python
# Cell 3: Batter Features vs FG
print('=== Batter Features vs FanGraphs ===')

batter_feat_fg_merge['fg_GB_pct'] = batter_feat_fg_merge['GB%'].apply(parse_fg_pct)
batter_feat_fg_merge['fg_LD_pct'] = batter_feat_fg_merge['LD%'].apply(parse_fg_pct)
batter_feat_fg_merge['fg_HR_FB'] = batter_feat_fg_merge['HR/FB'].apply(parse_fg_pct)

bat_feat_pairs = [
    ('wOBA', 'wOBA_fg', 'wOBA'),
    ('wRC_plus', 'wRC+', 'wRC_plus'),
    ('RE24', 'RE24_fg', 'RE24'),
    ('GB_pct', 'fg_GB_pct', 'GB_pct'),
    ('LD_pct', 'fg_LD_pct', 'LD_pct'),
    ('HR_FB_pct', 'fg_HR_FB', 'HR_FB_pct (expect FAIL)'),
]

bat_feat_summary = compare_stats(batter_feat_fg_merge, bat_feat_pairs[:2] + bat_feat_pairs[3:5], 'rate', 'bat_feat_vs_FG')
# index-type stats
bat_feat_idx = compare_stats(batter_feat_fg_merge, [bat_feat_pairs[2]], 'index', 'bat_feat_vs_FG')
# HR/FB separately — expected to fail
bat_feat_hrfb = compare_stats(batter_feat_fg_merge, [bat_feat_pairs[5]], 'rate', 'bat_feat_vs_FG')
```

```python
# Cell 4: Pitcher Season Stats vs FG
print('=== Pitcher Season Stats vs FanGraphs ===')

pitcher_fg_merge['fg_K_pct'] = pitcher_fg_merge['K%'].apply(parse_fg_pct)
pitcher_fg_merge['fg_BB_pct'] = pitcher_fg_merge['BB%'].apply(parse_fg_pct)
pitcher_fg_merge['fg_LOB_pct'] = pitcher_fg_merge['LOB%'].apply(parse_fg_pct)

pit_season_rate_pairs = [
    ('ERA', 'ERA_fg', 'ERA'),
    ('WHIP', 'WHIP_fg', 'WHIP'),
    ('K_pct', 'fg_K_pct', 'pitcher K_pct'),
    ('BB_pct', 'fg_BB_pct', 'pitcher BB_pct'),
    ('FIP', 'FIP_fg', 'FIP'),
    ('BABIP_allowed', 'BABIP', 'pitcher BABIP'),
    ('LOB_pct', 'fg_LOB_pct', 'LOB_pct'),
]
pit_season_idx_pairs = [
    ('K9', 'K/9', 'K/9'),
    ('BB9', 'BB/9', 'BB/9'),
    ('HR9', 'HR/9', 'HR/9'),
    ('ERA_minus', 'ERA-', 'ERA_minus'),
    ('FIP_minus', 'FIP-', 'FIP_minus'),
]

pit_season_summary = compare_stats(pitcher_fg_merge, pit_season_rate_pairs, 'rate', 'pit_season_vs_FG')
pit_season_idx_summary = compare_stats(pitcher_fg_merge, pit_season_idx_pairs, 'index', 'pit_season_vs_FG')
```

```python
# Cell 5: Pitcher Features vs FG
print('=== Pitcher Features vs FanGraphs ===')

pitcher_feat_fg_merge['fg_SwStr_pct'] = pitcher_feat_fg_merge['SwStr%'].apply(parse_fg_pct)
pitcher_feat_fg_merge['fg_CSW_pct'] = pitcher_feat_fg_merge['CSW%'].apply(parse_fg_pct)
pitcher_feat_fg_merge['fg_Zone_pct'] = pitcher_feat_fg_merge['Zone%'].apply(parse_fg_pct)
pitcher_feat_fg_merge['fg_Chase_pct'] = pitcher_feat_fg_merge['O-Swing%'].apply(parse_fg_pct)

pit_feat_rate_pairs = [
    ('xFIP', 'xFIP_fg', 'xFIP'),
    ('swstr_pct', 'fg_SwStr_pct', 'SwStr_pct'),
    ('csw_pct', 'fg_CSW_pct', 'CSW_pct'),
    ('zone_pct', 'fg_Zone_pct', 'Zone_pct'),
    ('chase_pct', 'fg_Chase_pct', 'Chase_pct'),
]
pit_feat_idx_pairs = [
    ('xFIP_minus', 'xFIP-', 'xFIP_minus'),
]

pit_feat_summary = compare_stats(pitcher_feat_fg_merge, pit_feat_rate_pairs, 'rate', 'pit_feat_vs_FG')
pit_feat_idx_summary = compare_stats(pitcher_feat_fg_merge, pit_feat_idx_pairs, 'index', 'pit_feat_vs_FG')
```

```python
# Cell 6: Pitcher Features vs Statcast LB
print('=== Pitcher Features vs Statcast Leaderboard ===')

sc_pairs = [
    ('xera', 'xera_sc', 'xERA'),
    ('brl_percent_allowed', 'brl_percent', 'brl_percent'),
    ('ev95percent_against', 'ev95percent', 'ev95percent'),
]

sc_summary = compare_stats(pitcher_feat_sc_merge, sc_pairs, 'rate', 'pit_feat_vs_Statcast')
```

```python
# Cell 7: Scatter Plots
os.makedirs('outputs/figures/eda_validation', exist_ok=True)

# Collect all comparison pairs with their merge DataFrames
all_plot_pairs = []
for pairs, df in [
    (bat_season_pairs, batter_fg_merge),
    (bat_feat_pairs, batter_feat_fg_merge),
    (pit_season_rate_pairs + pit_season_idx_pairs, pitcher_fg_merge),
    (pit_feat_rate_pairs + pit_feat_idx_pairs, pitcher_feat_fg_merge),
    (sc_pairs, pitcher_feat_sc_merge),
]:
    for ours_col, truth_col, label in pairs:
        all_plot_pairs.append((ours_col, truth_col, label, df))

# Individual scatter plots
for ours_col, truth_col, label, df in all_plot_pairs:
    sub = df[[ours_col, truth_col, 'year']].dropna()
    if len(sub) == 0:
        continue
    fig, ax = plt.subplots(figsize=(5, 5))
    scatter = ax.scatter(sub[ours_col], sub[truth_col], c=sub['year'], cmap='viridis',
                         alpha=0.4, s=10, edgecolors='none')
    lo = min(sub[ours_col].min(), sub[truth_col].min())
    hi = max(sub[ours_col].max(), sub[truth_col].max())
    pad = (hi - lo) * 0.05
    ax.plot([lo - pad, hi + pad], [lo - pad, hi + pad], 'r--', lw=1, label='y=x')
    ax.set_xlabel(f'Ours ({ours_col})')
    ax.set_ylabel(f'Truth ({truth_col})')
    ax.set_title(label)
    plt.colorbar(scatter, ax=ax, label='Year')
    fig.tight_layout()
    safe_label = label.replace('/', '_').replace(' ', '_').replace('(', '').replace(')', '')
    fig.savefig(f'outputs/figures/eda_validation/scatter_{safe_label}.png', dpi=100)
    plt.close(fig)

# Combined grid
n_plots = len(all_plot_pairs)
ncols = 5
nrows = (n_plots + ncols - 1) // ncols
fig, axes = plt.subplots(nrows, ncols, figsize=(ncols * 4, nrows * 4))
axes_flat = axes.flatten()

for i, (ours_col, truth_col, label, df) in enumerate(all_plot_pairs):
    ax = axes_flat[i]
    sub = df[[ours_col, truth_col, 'year']].dropna()
    if len(sub) == 0:
        ax.set_title(f'{label}\nNO DATA')
        continue
    ax.scatter(sub[ours_col], sub[truth_col], c=sub['year'], cmap='viridis',
               alpha=0.3, s=5, edgecolors='none')
    lo = min(sub[ours_col].min(), sub[truth_col].min())
    hi = max(sub[ours_col].max(), sub[truth_col].max())
    pad = (hi - lo) * 0.05
    ax.plot([lo - pad, hi + pad], [lo - pad, hi + pad], 'r--', lw=0.8)
    ax.set_title(label, fontsize=9)
    ax.set_xlabel('Ours', fontsize=7)
    ax.set_ylabel('Truth', fontsize=7)
    ax.tick_params(labelsize=6)

# Hide unused axes
for j in range(n_plots, len(axes_flat)):
    axes_flat[j].set_visible(False)

fig.suptitle('EDA Validation: Ours vs Truth', fontsize=14, y=1.01)
fig.tight_layout()
fig.savefig('outputs/figures/eda_validation/scatter_grid_all.png', dpi=150, bbox_inches='tight')
plt.close(fig)

print(f'Saved {n_plots} individual scatter plots + 1 combined grid to outputs/figures/eda_validation/')
```

```python
# Cell 8: Summary Table
print('=== VALIDATION SUMMARY ===')

summary = pd.concat(ALL_SUMMARIES, ignore_index=True)

display_cols = ['stat', 'source', 'mean_diff', 'mean_abs_diff', 'max_abs_diff',
                'correlation', 'n_matched', 'n_flagged', 'pass_fail']
summary = summary[display_cols]

# Style: highlight failures
def highlight_fail(row):
    if row['pass_fail'] == 'FAIL':
        return ['background-color: #ffcccc'] * len(row)
    elif row['pass_fail'] == 'PASS':
        return ['background-color: #ccffcc'] * len(row)
    return [''] * len(row)

styled = summary.style.apply(highlight_fail, axis=1)
display(styled)

n_pass = (summary['pass_fail'] == 'PASS').sum()
n_fail = (summary['pass_fail'] == 'FAIL').sum()
n_nodata = (summary['pass_fail'] == 'NO DATA').sum()
total = len(summary)

print(f'\nResults: {n_pass}/{total} PASS, {n_fail}/{total} FAIL, {n_nodata}/{total} NO DATA')

if n_fail > 0:
    print('\nFailed stats:')
    for _, row in summary[summary['pass_fail'] == 'FAIL'].iterrows():
        print(f"  - {row['stat']} ({row['source']}): mean_abs_diff={row['mean_abs_diff']}, corr={row['correlation']}")

print(f'\nVERDICT: {"ALL PASS" if n_fail == 0 else f"{n_fail} STAT(S) FAILED — review needed"}')
```
